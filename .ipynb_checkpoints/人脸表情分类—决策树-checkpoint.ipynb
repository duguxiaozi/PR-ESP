```python
import numpy as np
import pandas as pd
```

## 1、导入数据

```python
Xdata = np.load('../Xtrain.npy')
Xdata.shape

Xlabel = np.load('../Xlabel.npy')
pd.DataFrame(Xlabel)
```

## 2、目标建立数组

```python
Xmood = Xlabel[:,[8]]
Xmoodlabel=np.ones((3991,1))

for i in range(3991):
    a=int(Xmood [i,:])
    Xmoodlabel[i,:]=a

pd.DataFrame(Xmoodlabel)
```

## 3、图片数据转换为同一行中

```python
X=np.ones((3991,16384))

for i in range(3991):
    a=Xdata[i].flatten()
    X[i,:]=a

X.shape
Xtrain=X[0:3550,:]
Ytrain=Xmoodlabel[0:3550,:]
Xtest=X[3551:3990,:]
Ytest=Xmoodlabel[3551:3990,:]
```

## 4、图片数据标准化

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_Std_train = sc.fit_transform(Xtrain)
X_Std_test = sc.fit_transform(Xtest)

X_Std_train.std()
```

## 5、决策树模型建立以及判断过拟合结果

```python
from sklearn import tree
```

```python
#训练
clf = tree.DecisionTreeClassifier(criterion="entropy",splitter="random",max_depth=6)
clf.fit(X_Std_train, Ytrain.ravel())

#测试
score = clf.score(X_Std_train,Ytrain.ravel())
print('预测准确率：',score)
```

## 6、随机森林模型建立以及判断过拟合结果

```python
# Fitting Random Forest Classification to the Training set
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_Std_train, Ytrain.ravel())

#测试
score = rfc.score(X_Std_train,Ytrain.ravel())
print('预测准确率：',score)
```

## 7. 比较单一的决策树和随机森林的训练集结果

```python
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(X,Xmoodlabel,test_size=0.3)

clf = DecisionTreeClassifier(random_state=0)
rfc = RandomForestClassifier(random_state=0)
clf = clf.fit(xtrain,ytrain.ravel())
rfc = rfc.fit(xtrain,ytrain.ravel())
score_c = clf.score(xtest,ytest.ravel())
score_r = rfc.score(xtest,ytest.ravel())
print("Single Tree:{}".format(score_c)
     ,"Random Forest:{}".format(score_r)
     )
```

## 8. 在CV=3交叉验证下决策树和随机森林的效果对比

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
clf_1 = []
rfc_1 = []

for i in range(10):
    rfc = RandomForestClassifier(n_estimators=10)
    rfc_s = cross_val_score(rfc,X,Xmoodlabel.ravel(),cv=3).mean()
    rfc_1.append(rfc_s)
    
    clf = DecisionTreeClassifier()
    clf_s = cross_val_score(clf,X,Xmoodlabel.ravel(),cv=3).mean()
    clf_1.append(clf_s)
    
plt.plot(range(1,11),rfc_1,label = "Random Forest")
plt.plot(range(1,11),clf_1,label = "Decision Forest")
plt.legend()
plt.show()
```

## 9、随机森林参数n_estimators的学习曲线

```python
superpa = []
for i in range(100):
    rfc = RandomForestClassifier(n_estimators=i+1,n_jobs=-1)
    rfc_s = cross_val_score(rfc,X,Xmoodlabel.ravel(),cv=3).mean()
    superpa.append(rfc_s)
print(max(superpa),superpa.index(max(superpa)))
plt.figure(figsize=[20,5])
plt.plot(range(1,101),superpa)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 87)
rfc.fit(X_Std_train, Ytrain.ravel())

#测试
score = rfc.score(X_Std_test,Ytest.ravel())
print('预测准确率：',score)
scores = cross_val_score(clf, X_Std_train , Ytrain.ravel(), cv=5, scoring='accuracy')
print('随机森林分类器',scores.mean())
```

## 10、PCA降维

### 10.1选择最好的n_components:累积可解析方差贡献率曲线

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
pca_line = PCA().fit(X)
plt.plot(np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks()
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance ratio")
plt.show()
```

### 10.1 PCA降维后的决策树准确率图像

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

Pca = []
for i in range(100):
    pca = PCA(n_components=i+1)
    X_pca = pca.fit_transform(X)
    xtrain_pca,xtest_pca,ytrain,ytest=train_test_split(X_pca,Xmoodlabel,test_size=0.3)
    #train
    clf = tree.DecisionTreeClassifier(criterion="entropy",splitter="random",max_depth=6)
    clf.fit(xtrain_pca, ytrain.ravel())
    #test
    clf_s =clf.score(xtest_pca,ytest.ravel())
    Pca.append(clf_s)
print(max(Pca),Pca.index(max(Pca)))
plt.figure(figsize=[20,5])
plt.plot(range(1,101),Pca)
plt.show()
```

### 10.2 PCA降维后的随机森林准确率图像

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

Pca1 = []
for i in range(100):
    pca = PCA(n_components=i+1)
    X_pca = pca.fit_transform(X)
    xtrain_pca,xtest_pca,ytrain,ytest=train_test_split(X_pca,Xmoodlabel,test_size=0.3)
    #train
    rfc = RandomForestClassifier()
    rfc.fit(xtrain_pca, ytrain.ravel())
    #test
    rfc_s =rfc.score(xtest_pca,ytest.ravel())
    Pca1.append(rfc_s)
print(max(Pca1),Pca1.index(max(Pca1)))
plt.figure(figsize=[20,5])
plt.plot(range(1,101),Pca1)
plt.show()
```
